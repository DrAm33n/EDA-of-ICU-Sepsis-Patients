**<h2 style="color:purple;">Processing Missing Values and Column Wrangling**
</h2>

```python
# removed columns Variable,Patient Accomodation,Others and Comorbidity Names
df=df.drop(columns={'Variable','Patient_Accommodation','Others','Comorbidity_Names'})

#droping first row which contained only null values
df=df.drop(0)

#Droping rows from 2939 to 4520 which conatined only null values using range 
df=df.drop(range(2939,4521))
```

**<h2 style="color:purple;">Filling Missing Values**
</h2>

```python
# Replacing the null values with mode
df_mode=df.iloc[:,3:11].mode().iloc[0] 
df.iloc[:,3:11] = df.iloc[:,3:11].fillna(df_mode) 

df_mod = df['History_of_hospitalization_in_the_last_3_months'].mode().iloc[0]
df['History_of_hospitalization_in_the_last_3_months'] = df['History_of_hospitalization_in_the_last_3_months'].fillna(df_mod)

# Replacing the null values with median
df_med = df.iloc[:,11:13].median()
df.iloc[:,11:13]=df.iloc[:,11:13].fillna(df_med)

# Replacing the null values with mean
df_mean = df_mean = df.iloc[:,17:33].mean()
df.iloc[:,17:33] = df.iloc[:,17:33].fillna(df_mean)

# Replacing the null values with 0 as it is more logical
df.iloc[:,-4:] = df.iloc[:,-4:].fillna('0')

df['Reproduction_in_Blood_Culture'] = df['Reproduction_in_Blood_Culture'].fillna(0)
```

```python
# Checking for null values after cleaning
df.isnull().sum()
```

**<h2 style="color:purple;">Column Wrangling**
</h2>

```python
# Changing column names to lowercase
df.columns = df.columns.str.lower().str.strip()

# Changing column records in resistance_status to propercase
df['resistance_status'] = df['resistance_status'].str.strip('.').str.title()
```

```python
# renaming column for better anlysis
cols_to_rename={
    'the_national_early_warning_score_news' : 'news',
    'nt_pro_bnp_pro_brain_natriuretic_peptide' : 'nt_probnp',
    'systemic_inflammatory_response_syndrome_sirs_presence':'sirs',
    'if_reproduction_is_present_in_blood_culture_factor' : 'positive_organisms'}

df = df.rename(columns=cols_to_rename)
```

```python
# Conversion of datatypes
col_type = ['positive_organisms',  # Assigning the columns to variable
            'resistance_status',
            'culture_type',
           'antibiotherapy']

df[col_type] = df[col_type].apply(lambda x: x.astype("string").str.strip()) # using lambda function to convert the columns
```

```python
# Replacing the values for better analysis

df['gender'] = df['gender'].replace({1 : 'M' , 2 : 'F'}).astype('string') #replacing 1 and 2 to male and female

df['diabetes_mellitus'] = df['diabetes_mellitus'].replace({10:1}) # Replacing 10 to 1 in the diabetes_mellitus column

df['mortality'] = df['mortality'].replace({1: 'Expired', 0: 'Alive', 3:'Alive'}).astype('string') #replacing 0 and 1 to Alive and Expired

df['positive_organisms'] = df['positive_organisms'].replace('0','No growth') #replacing 0 to no growth for better outcome

df['resistance_status'] = df['resistance_status'].replace('0','No resistance') #replacing 0 to no resistance for better analysis

df['culture_type'] = df['culture_type'].replace('0','Other Types') #replacing 0 to other types for better analysis

df['antibiotherapy'] = df['antibiotherapy'].replace('0','Other Antibiotics') #replacing 0 to other antibiotics for better analysis
```

```python
# Derivation and dropping of columns

# Deriving nlr column from neutrophil_count and lymphocyte_count 
df['nlr'] = (df['neutrophil_count'] / df['lymphocyte_count']).replace([np.inf, -np.inf], np.nan).fillna(0) # to avid infinite and null value

# Deriving mlr column from monocyte_count and lymphocyte_count 
df['mlr'] = (df['monocyte_count'] / df['lymphocyte_count']).replace([np.inf, -np.inf], np.nan).fillna(0)  # to avid infinite and null value

#Deriving mean arterial pressure from systolic pressure and diastolic pressure
df['map'] = (df['systolic_blood_pressure'] + 2*df['diastolic_blood_pressure']) / 3 

# Dropping the columns which are not necessary after derivation
df = df.drop( columns = ['neutrophil_count','lymphocyte_count','monocyte_count','eosinophil_count'])
```

**<h3 style="color:darkgreen;">2.1 Average Length of Stay of each Age Group**

**Goal:** To analyze the relationship between patient age groups and the average length of ICU stay among sepsis patients.

```python
# Generating age groups for esier analysis
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20','21-40','41-60','61-79','80+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, ordered=True)

# Calculate mean ICU stay
age_mean = df.groupby('age_group')['length_of_stay_in_intensive_care'].mean().reset_index()

# Plotly bar chart
fig = px.bar(
    age_mean,
    x='age_group',
    y='length_of_stay_in_intensive_care',
    title='Average ICU Length of Stay by Age Group',
    labels={
        'age_group':'Age Group',
        'length_of_stay_in_intensive_care':'Average ICU Stay'
    }
)

# Configure the figure size
fig.update_layout(
    width=1000,
    height=700
)

# Displaying the Visualization
fig.show()
```

**<h4 style="color:darkred;"> Insights</h4>**

- The bar chart shows that the average ICU length of stay is higher among patients in the 60–79 and 80+ age groups compared to younger age groups.
- This suggests that older patients may require longer hospitalization, possibly due to reduced physiological reserve and a higher prevalence of underlying comorbidities.
- The findings indicate that age may influence disease progression and recovery time in sepsis patients, with elderly patients potentially experiencing more severe clinical conditions.
- It also suggests that treatment outcomes and recovery duration may vary depending on the patient’s age group, highlighting the importance of age as a factor in sepsis management.

**<h3 style="color:darkgreen;">2.5 Mean Arterial Pressure VS Mortality**
</h3>

**Goal:** To analyze the relationship between Mean Arterial Pressure (MAP) and mortality outcome in sepsis patients.

```python
# Computing the average MAP
bp_mean = df.groupby('mortality')['map'].mean()
print(bp_mean)

# Set the figure size for clarity
plt.figure(figsize=(6,5))

# Plotting line chart to identify the relationship
plt.plot(bp_mean.index, bp_mean.values, marker='o', linewidth=2, label='Mean Arterial Pressure')

# Plotting the critical MAP threshold used in sepsis management
plt.axhline(65, linestyle='--', color='red', label='Critical MAP Threshold (65 mmHg)')

plt.title('Average Mean Arterial Pressure by Mortality Outcome')
plt.xlabel('Mortality Outcome')
plt.ylabel('Average MAP')

plt.grid(alpha=0.3)
plt.legend()

# Displaying the Visualization
plt.show()
```

**<h4 style="color:darkred;"> Insights</h4>**

- The line chart compares the average Mean Arterial Pressure (MAP) between patients who survived and those who did not survive.
- In sepsis management, maintaining a MAP of at least 65 mmHg is generally recommended to ensure adequate organ perfusion and to reduce the risk of organ dysfunction.
- In this analysis, patients who expired had an average MAP of approximately 71 mmHg, which is above the commonly recommended threshold.
- This observation suggests that although maintaining an adequate MAP is important for preventing organ hypoperfusion, mortality in sepsis can still occur due to other factors such as severe infection, organ failure, comorbidities, or delayed treatment.

**<h2 style="color:purple;">3. Multivariate Anlaysis**

**<h3 style="color:darkgreen;">3.1 Average ICU Stay by Resistance Status and Antibiotherapy**
</h3>

**Goal:** To analyze whether antibiotic-resistant infections are associated with increased hospital stay in sepsis patients.

```python
# Grouping the top 10 anibiotics
top_ab = df['antibiotherapy'].value_counts().head(10).index
df_ab = df[df['antibiotherapy'].isin(top_ab)]

# Pivot table for easier analysis 
pivot = df_ab.pivot_table(
    values='length_of_stay_in_intensive_care',
    index='resistance_status',
    columns='antibiotherapy',
    aggfunc='mean'
)

# Set the figure size for clarity
plt.figure(figsize=(10,10))

# Plotting heatmap
sns.heatmap(pivot, annot=True, fmt='.1f', cmap='Blues')

plt.title('Average ICU Stay by Resistance Status and Antibiotic Therapy')
plt.xlabel('Antibiotic Therapy')
plt.ylabel('Resistance Status')

# Displaying the Visualization
plt.show()
```

**<h4 style="color:darkred;"> Insights</h4>**

- The heatmap illustrates the average ICU length of stay across different antibiotic resistance statuses and antibiotic therapies.
- Patients with certain resistant organisms, such as inducible beta-lactamase–producing organisms, appear to have longer ICU stays, particularly when treated with specific antibiotics like gentamicin.
- In contrast, patients with no resistance or methicillin-sensitive organisms generally show shorter ICU stays, suggesting that infections caused by susceptible pathogens may be easier to manage clinically.
- Some antibiotic–resistance combinations show very high ICU stay values, which may indicate more complicated infections, delayed treatment response, or severe disease conditions.
- The analysis suggests that antimicrobial resistance may contribute to prolonged hospitalization in sepsis patients

</h3>

**Goal:** To analyze whether the presence of multiple comorbidities in sepsis patients is associated with an increase in mortality rate and longer ICU length of stay.

```python
# Assigning the columns to a variable
comorb_cols = [
    'solid_organ_cancer',
    'hematological_diseases',
    'hypertension',
    'heart_diseases',
    'diabetes_mellitus',
    'chronic_renal_failure',
    'neurological_diseases',
    'copd_asthma'
]

# Count comorbidities
df['comorbidity_count'] = df[comorb_cols].sum(axis=1)

# Computing the Mortality rate
mortality_rate = df.groupby('comorbidity_count')['mortality'].apply(
    lambda x: (x == 'Expired').mean()
)

# Computing the average length of stay
los_mean = df.groupby('comorbidity_count')['length_of_stay_in_intensive_care'].mean()

# Generating dataframe by Combining the Results
df_analysis = pd.DataFrame({
    'Mortality Rate': mortality_rate,
    'Average ICU Stay': los_mean
}).reset_index()

# Generating Canvas
fig = go.Figure()

# Mortality Rate (Bar)
fig.add_trace(
    go.Bar(
        x=df_analysis['comorbidity_count'],
        y=df_analysis['Mortality Rate'],
        name='Mortality Rate'
    )
)

# ICU Length of Stay (Line)
fig.add_trace(
    go.Scatter(
        x=df_analysis['comorbidity_count'],
        y=df_analysis['Average ICU Stay'],
        name='Average ICU Stay',
        yaxis='y2',
        mode='lines+markers'
    )
)

# Configuring The Layout
fig.update_layout(
    title='Comorbidity Burden vs Mortality and ICU Length of Stay',
    xaxis_title='Number of Comorbidities',
    yaxis=dict(title='Mortality Rate'),
    yaxis2=dict(
        title='Average ICU Length of Stay',
        overlaying='y',
        side='right'
    ),
# Adjusting Chart Size    
    width=1000,
    height=900
)

# Displaying the Visualization
fig.show()
```

**<h4 style="color:darkred;"> Insights</h4>**

- Patients with more than three comorbidities show a higher mortality rate, indicating that a greater burden of underlying diseases may increase the risk of poor outcomes in sepsis.
- Interestingly, patients without comorbidities also demonstrate a relatively high mortality rate compared to those with one or two comorbid conditions.
- This suggests that mortality does not necessarily increase linearly with the number of comorbidities, and other clinical factors may also influence patient outcomes.
- The average ICU length of stay increases with the number of comorbidities, indicating that patients with multiple underlying diseases often require longer and more complex medical management.
- However, patients with more than three comorbidities show a shorter length of stay despite having a higher mortality rate, which may suggest rapid clinical deterioration and reduced response to treatment.
- Additionally, patients without comorbidities have an average ICU stay of approximately 8 days but still exhibit a higher mortality rate, which may indicate the presence of severe acute infection or delayed diagnosis in these cases.
